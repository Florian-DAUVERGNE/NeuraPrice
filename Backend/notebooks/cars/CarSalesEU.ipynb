```python
dataset.drop(['Date_premiere_mise_en_circulation', 'Prix_neuf', 'Annee_modele', 'Modele'], axis = 1, inplace=True)
```

```python
dataset.info()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df = dataset
```

```python
df['Marque'] = le.fit_transform(df['Marque'])
df['Carburant'] = le.fit_transform(df['Carburant'])
df['Boite_vitesse'] = le.fit_transform(df['Boite_vitesse'])
df['Critere_air'] = le.fit_transform(df['Critere_air'])
df['Km_bins'] = le.fit_transform(df['Km_bins'])
df['Couleur'] = le.fit_transform(df['Couleur'])
df['Type_vehicule'] = le.fit_transform(df['Type_vehicule'])
df['Pays'] = le.fit_transform(df['Pays'])
```

```python
df.info()
```

```python
X = df.drop(['Prix_occassion'],axis=1)
Y = df['Prix_occassion']
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVR
from sklearn import linear_model
from sklearn.metrics import (confusion_matrix,accuracy_score,recall_score,precision_score, mean_absolute_error, r2_score, mean_squared_error)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(x_train.shape, x_test.shape,  y_train.shape, y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(x_train, y_train)
reg.score(x_train, y_train)
```

```python
regr = RandomForestRegressor(max_depth=110, n_estimators=210, random_state=0)
regr.fit(x_train, y_train)
```

```python
y_pred = regr.predict(x_test)
score = r2_score(y_test, y_pred)
round(score, 2)
```

```python
models = {
  'LinearRegression' : linear_model.LinearRegression(),
  'RandomForestRegressor' : RandomForestRegressor(),
  'SVR' : SVR()
}
```

```python
scores_df = pd.DataFrame()
for model_name, model in models.items() :
  print('#'*50)
  print('model en cours :', model_name)
  model.fit(x_train, y_train)
  predictions = model.predict(x_test)
  scores_df = pd.concat([
      scores_df,
      pd.DataFrame([{
          'model' : model_name,
          'R2' : r2_score(y_test, predictions),
          'MAE' : mean_absolute_error(y_test, predictions),
          'MSE' : mean_squared_error(y_test, predictions)
      }])
  ], axis=0)
  plt.figure(figsize=(20, 10))
  plt.scatter(x=range(30), y=y_test[:30], label='Real')
  plt.scatter(x=range(30), y=predictions[:30], label='Preds')
  plt.legend()
  plt.show()
```

```python
scores_df.sort_values('R2', ascending=False)
```

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, ExtraTreesRegressor, HistGradientBoostingRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor

models = {
    "RandomForest": RandomForestRegressor(max_depth=110, n_estimators=210, random_state=0),
    "GradientBoosting": GradientBoostingRegressor(),
    "XGBoost": XGBRegressor(),
    "LightGBM": LGBMRegressor(),
    "ExtraTrees": ExtraTreesRegressor(),
    "HistGradientBoosting": HistGradientBoostingRegressor()
}

# Initialiser les variables pour suivre le meilleur modle
best_model = None
best_model_name = ""
best_r2 = float("-inf")
best_metrics = {}
results = []

for name, model in models.items():
  print(f"Entranement du modle {name}...")
  model.fit(x_train, y_train)
  y_pred = model.predict(x_test)

  r2 = r2_score(y_test, y_pred)
  mae = mean_absolute_error(y_test, y_pred)
  mse = mean_squared_error(y_test, y_pred)
  print(f"{name} - R2: {r2:.4f}, MAE: {mae:.4f}, MSE: {mse:.4e}")

  results.append([name, r2, mae, mse])
  # Mettre  jour le meilleur modle si ncessaire
  if r2 > best_r2:
    best_model = model
    best_model_name = name
    best_r2 = r2
    best_metrics = {"R2": r2, "MAE": mae, "MSE": mse}

results_df = pd.DataFrame(results, columns=["Model", "R2", "MAE", "MSE"])
print(results_df.sort_values(by="R2", ascending=False))

# Afficher le meilleur modle
print(f"\nLe meilleur modle est {best_model_name} avec R2 = {best_r2:.4f}")
print(f"Mtriques du meilleur modle: {best_metrics}")

```
