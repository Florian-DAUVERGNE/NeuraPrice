## ***Outliers / Underliers***

```python
def remove_outliers_std_dataframe(df, threshold=3):
    """
    Removes outliers using the standard deviation for all columns in a DataFrame.

    Parameters:
    df (pd.DataFrame): The DataFrame to be processed.
    threshold (float): The number of standard deviations above which to remove outliers.

    Returns:
    pd.DataFrame: The DataFrame without the outliers.
    """
    # Apply outlier suppression to each column
    for column in df.select_dtypes(include=['float64', 'int64']).columns:
        mean_val = df[column].mean()
        std_val = df[column].std()

        # Calculate limits
        lower_limit = mean_val - threshold * std_val
        upper_limit = mean_val + threshold * std_val

        # Delete outliers / underliers
        df = df[(df[column] >= lower_limit) & (df[column] <= upper_limit)]

    return df
```

```python
def separate_cat_num_variable(X):
    """
    Separates features of a DataFrame into numeric and categorical types.

    Parameters:
    X (pd.DataFrame): The DataFrame containing features.

    Returns:
    tuple: A tuple containing lists of numeric and categorical feature names.
    """
    numeric_features = []
    categorical_features = []
    bool_features = []

    for i, t in X.dtypes.items():
        if ('float' in str(t)) or ('int32' in str(t)) or ('int64' in str(t)):
            numeric_features.append(i)
        
        elif ('bool' in str(t)):
            bool_features.append(i)
            
        else:
            categorical_features.append(i)
            
    categorical_features.pop(1)
    categorical_ordinal_features = ["City"]
    numeric_features.remove("price_per_sq_m")
    
    print('Found numeric features:', numeric_features)
    print('Found categorical features:', categorical_ordinal_features)
    print('Found bool features:', bool_features)
    print('Found cat features:', categorical_features)

    return numeric_features, categorical_features, bool_features, categorical_ordinal_features
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import  OneHotEncoder, StandardScaler, LabelEncoder, OrdinalEncoder, FunctionTransformer
from sklearn.compose import ColumnTransformer
```

```python
# Fonction explicite pour convertir les booléens en entiers (évite lambda => problème avec l'enregistrement du pipeline)
def convert_bool_to_int(X):
    return X.astype(int)

def Create_pipeline_preprocessor(numeric_features, categorical_features, bool_features, categorical_ordinal_features):
    """
    Create a pipeline for preprocessing features.

    Parameters:
    - numeric_features : list of numeric columns
    - categorical_features : list of categorical features
    - bool_features : list of boolean features
    - categorical_ordinal_features : list of ordinal categorical features

    Returns:
    - preprocessor : ColumnTransformer object that applies transformations to the dataset
    """

    # Pipeline for numeric features (StandardScaler for normalization)
    numeric_transformer = StandardScaler()

    # Pipeline for ordinal categorical features (OrdinalEncoder)
    categorical_ordinal_transformer = OrdinalEncoder()

    # Pipeline for nominal categorical features (OneHotEncoder, dropping first to avoid collinearity)
    categorical_transformer = OneHotEncoder(drop='first', handle_unknown='ignore')

    # Pipeline for boolean features (convert bool to int using a function instead of lambda)
    bool_transformer = FunctionTransformer(convert_bool_to_int)

    # ColumnTransformer to apply transformations
    preprocessor = ColumnTransformer(
        transformers=[
            ('num', numeric_transformer, numeric_features),
            ('cat_ord', categorical_ordinal_transformer, categorical_ordinal_features),
            ('cat', categorical_transformer, categorical_features),
            ('bool', bool_transformer, bool_features)
        ]
    )

    return preprocessor
```

```python
preprocessor = Create_pipeline_preprocessor(numeric_features, categorical_features, bool_features, categorical_ordinal_features)
```

```python
import pickle
import dill

def apply_preprocessing(preprocessor, X_train, Y_train, X_test, Y_test, save_path="../models/preprocessor_real_estate.pkl"):
    """
    Apply preprocessing steps on training and test sets.

    Parameters:
    - preprocessor (object): The preprocessor (e.g., ColumnTransformer) for feature transformations.
    - X_train (DataFrame or ndarray): Training features.
    - Y_train (Series or ndarray): Training labels.
    - X_test (DataFrame or ndarray): Test features.
    - Y_test (Series or ndarray): Test labels.
    - save_path (str): Path where the fitted preprocessor should be saved.

    Returns:
    - X_train_transformed (ndarray): Processed training features.
    - Y_train (ndarray): Encoded training labels.
    - X_test_transformed (ndarray): Processed test features.
    - Y_test (ndarray): Encoded test labels.
    """

    print("\n🔹 Entraînement du préprocesseur sur le train set...")

    # Entraîner le préprocesseur sur X_train uniquement (PAS `fit_transform()`)
    preprocessor.fit(X_train)

    # sauvegarde du préprocesseur entraîné
    with open(save_path, "wb") as f:
        dill.dump(preprocessor, f) # utiliser dill pour garder les fonctions personnalisées dans le pickle
    print(f"✅ Préprocesseur entraîné et sauvegardé dans {save_path}")

    # Appliquer la transformation sur X_train
    print("\n🔹 Transformation de X_train...")
    X_train_transformed = preprocessor.transform(X_train)
    print(f"✅ X_train transformé. Shape: {X_train_transformed.shape}")

    # Vérification Y_train
    Y_train = np.array(Y_train)
    print(f"✅ Labels Y_train convertis en array. Shape: {Y_train.shape}")

    # Appliquer la transformation sur X_test (⚠️ Ne pas refit !)
    print("\n🔹 Transformation de X_test...")
    X_test_transformed = preprocessor.transform(X_test)
    print(f"✅ X_test transformé. Shape: {X_test_transformed.shape}")

    # Vérification Y_test
    Y_test = np.array(Y_test)
    print(f"✅ Labels Y_test convertis en array. Shape: {Y_test.shape}")

    print("\n✅ Preprocessing terminé avec succès !")

    return X_train_transformed, Y_train, X_test_transformed, Y_test
```

```python
# Apply function 
X_train_transformed, Y_train, X_test_transformed, Y_test = apply_preprocessing(preprocessor, X_train, Y_train, X_test, Y_test, save_path = "../models/preprocessor_real_estate.pkl")
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score, mean_squared_error
```

```python
# Enregistrer le Pipeline de transformation polynomial et de selction de variable sans le modèle
poly_select_pipeline_real_estate = Pipeline([
    ('polynomial', PolynomialFeatures(degree=2, include_bias=False)),
    ('select', SelectKBest(score_func=f_regression, k=12))
])

poly_select_pipeline_real_estate.fit(X_train_transformed, Y_train)

with open("../models/poly_select_pipeline_real_estate.pkl", "wb") as f:
    dill.dump(poly_select_pipeline_real_estate, f)
    print(f"pipeline sauvezgardé sur dans le chemin : ../models/poly_select_pipeline_real_estate.pkl")
    
# Transformation de X_train et X_test
X_train_selected = poly_select_pipeline_real_estate.transform(X_train_transformed)
X_test_selected = poly_select_pipeline_real_estate.transform(X_test_transformed)

print(f"✅ Transformation polynomiale et sélection de features terminées")
print(f"X_train_selected shape: {X_train_selected.shape}")
print(f"X_test_selected shape: {X_test_selected.shape}")
```

```python
estimator = RandomForestRegressor(random_state = 42)
```

```python
hyper_params = {
    'max_depth': [40],
    'min_samples_leaf': [3, 4],
    'min_samples_split': [2, 3],
    'n_estimators': [200]}
```

```python
from sklearn.metrics import make_scorer

# Define the metrics to evaluate
scoring_metrics = {
    'r2': make_scorer(r2_score),  # R2
    'mse': make_scorer(mean_squared_error, greater_is_better=False)  # MSE (minimisation)
}
```

```python
from sklearn.model_selection import GridSearchCV

# If refit is set to True, it means that once the grid search is complete
# and the best parameters are identified, the model is automatically fitted to the entire training dataset with these best parameters.
# This can be convenient as it gives you a ready-to-use model with the best parameters,
# and you can directly use it to make predictions on new data without manually adjusting the model with the best parameters.

# Perform Grid Search with Cross-Validation
gridsearch = GridSearchCV(estimator = estimator, scoring=scoring_metrics, param_grid=hyper_params, cv=10, refit='r2', verbose=2, n_jobs=-1)
gridsearch.fit(X_train_selected, Y_train)
print("Grid Search and Cross-Validation...Done.")

# Get the best model based on the grid search
best_model = gridsearch.best_estimator_

# Predictions on the training set
y_train_pred = best_model.predict(X_train_selected)
# Predictions on the test set
y_test_pred = best_model.predict(X_test_selected)

# Display evaluation metrics for the training set
print('Train:')
print(r2_score(Y_train, y_train_pred))
print("Best hyperparameters: ", gridsearch.best_params_)
print("Best validation base on r2 score: ", gridsearch.best_score_)
print(mean_squared_error(Y_train, y_train_pred))

print()

# Display evaluation metrics for the test set
print('Test:')
print(r2_score(Y_test, y_test_pred))
print("Best hyperparameters: ", gridsearch.best_params_)
print(mean_squared_error(Y_test, y_test_pred))
```
